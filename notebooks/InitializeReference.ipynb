```python
#import packages
import pandas as pd
import csv
import requests
import json

# set constants
coin_metrics_url = 'https://api.coinmetrics.io/v2'
coin_metrics_api_key = 'api_key=DlgTDcpDmRS8H5gNnila' #append to every request
```

```python
#generate metrics reference
coin_metrics_session = requests.Session()
url = coin_metrics_url+ '/metrics?' + coin_metrics_api_key
response = coin_metrics_session.get(url, verify=False)
available_metrics = json.loads(response.text)['metrics']
metric_reference = []
for i in range(len(available_metrics)):
    metric_url = coin_metrics_url+ '/metrics/' + available_metrics[i] + '?' + coin_metrics_api_key
    metric_response = coin_metrics_session.get(metric_url, verify=False)
    metric_info = json.loads(metric_response.text)['metricInfo']
    metric_reference.append({
        'id' : metric_info['id'],
        'name' : metric_info['name'],
        'category' : metric_info['category'],
        'subcategory' : metric_info['subcategory'],
        'description' : metric_info['description'],
    })
```

```python
#export to csv
keys = metric_reference[0].keys()
with open('reference.csv', 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(metric_reference)
```

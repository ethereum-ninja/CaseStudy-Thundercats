```python
import pandas as pd
import numpy as np
import pdblp as bbg
import datetime as dt
import matplotlib.pyplot as plt
bbcon = bbg.BCon(debug=False, port=8194, host='SAPI', timeout=5000)
bbcon.start()
```

```python
transaction_data = pd.read_csv('transaction_data.csv', parse_dates=['date'], date_parser=parse_date)

transaction_data = transaction_data[np.logical_or(
    transaction_data.payment_token_id == 1,
    transaction_data.payment_token_id == 2
)]
transaction_data['EthereumUnits'] =  transaction_data.total_price /10**18
start_date = transaction_data.date.min()
end_date = transaction_data.date.max()
prices = get_prices(dt.datetime.strftime(start_date, '%Y%m%d'),  dt.datetime.strftime(end_date, '%Y%m%d'))
#prices.date = parse_date(prices.date)
transaction_data = transaction_data.merge(prices, on=['date'])
transaction_data['dollars'] = transaction_data.EthereumUnits * transaction_data.value
transaction_data['week'] = transaction_data.date.dt.week
transaction_data['month'] =  transaction_data.date.dt.month

transaction_data.groupby('date')['dollars'].sum().to_csv('dollars_by_date.csv', header=True)
transaction_data.groupby('week')['dollars'].sum().to_csv('dollars_by_week.csv', header=True)
transaction_data.groupby('month')['dollars'].sum().to_csv('dollars_by_month.csv', header=True)
transaction_data.groupby('date')['dollars'].mean().to_csv('avg_dollars_by_date.csv', header=True)
transaction_data.groupby('week')['dollars'].mean().to_csv('avg_dollars_by_week.csv', header=True)
transaction_data.groupby('month')['dollars'].mean().to_csv('avg_dollars_by_month.csv', header=True)
transaction_data.groupby('date')['dollars'].count().to_csv('transaction_count_by_date.csv', header=True)
transaction_data.groupby('week')['dollars'].count().to_csv('transaction_count_by_week.csv', header=True)
transaction_data.groupby('month')['dollars'].count().to_csv('transaction_count_by_month.csv', header=True)


plt.show(transaction_data.groupby('date')['dollars'].sum().plot())
plt.show(transaction_data.groupby('week')['dollars'].sum().plot())
plt.show(transaction_data.groupby('month')['dollars'].sum().plot())

plt.show(transaction_data.groupby('date')['dollars'].mean().plot())
plt.show(transaction_data.groupby('week')['dollars'].mean().plot())
plt.show(transaction_data.groupby('month')['dollars'].mean().plot())

plt.show(transaction_data.groupby('date')['id'].count().plot())
plt.show(transaction_data.groupby('week')['id'].count().plot())
plt.show(transaction_data.groupby('month')['id'].count().plot())

summary = transaction_data.describe()[['EthereumUnits', 'value', 'dollars']]
print([transaction_data.EthereumUnits.median(), 
                               transaction_data.value.median(), 
                               transaction_data.dollars.median()])

summary.to_csv('transaction_summary.csv')
transaction_data.to_csv('transactions_with_value.csv')               
```

```python
def parse_date(date):
    return pd.to_datetime(date, format='%Y-%m-%d')
```

```python
def get_prices(start_date, end_date):
    return bbcon.bdh('XET Curncy', ['PX_LAST'], start_date, end_date, longdata=True)
```

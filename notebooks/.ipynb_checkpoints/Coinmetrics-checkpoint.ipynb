```python
def add_b_function(data, price_key):
    #do we have to sample daily ticks here? what is the concept of a tick in crypto?
    '''set b feature (tick imbalance function)
    b(t) = b(t-1) if price change is zero
    b(t) = abs(return)/return if change is non-zero
    Total Tick Imbalance at time T: Theta(T) = SUM b(t) from t = 1 to T
    Expected value of Theta(T) at t = 0
    E0T = Expected size of the tick bar 
        (calc'd as exponentially weighted moving average of of T values from prior bars)
    B0T = Expected Probabilty of up(?) tick at beginning of bar 
        (calc'd as exponentially weighted moving average of of b(t) values from prior bars)
        consider window range, probability = SUM b(s)/count of elements in set defining T window
    Find minimum value for T such that:
        |Theta(T)| >= E0T * B0T
    Create contiguous range of these values
    '''
    #step 1 - diff DF
    data['return'] = data[price_key].pct_change()
    last_b = [np.nan]
    def helper(row, last):
        b = last[0] if row['return'] == 0 else row['return']/abs(row['return'])
        last_b[0] = b
        return b
    data['b(t)'] = data.apply(lambda x: helper(x, last_b), axis=1)
    return data

```

```python
def theta_min(up_prob, expected_t):
    return math.ceil(expected_t * (2*up_prob - 1))
```

```python
def next_interval_bound(data, theta, index):
    return data[index:][np.abs(data['Theta(t)']) >= abs(theta)].index[0] 
```

```python
def up_tick_probability(data):
    return len(data[data['b(t)'] == 1])/data.shape[0]
```

```python
def hist_sub_plot(data, title=None):
    fig, ax = plt.subplots(1, len(data.columns), squeeze=False)
    ax = ax.ravel()
    i = 0
    for col in data.columns:
        ax[i].hist(data[col], bins='auto', alpha=0.5)
        if title:
            ax[i].set_title(title)
        i = i+1
    fig
```

```python
def numerical_normality_test(data, alpha = 0.05):
    #clean_data = data.replace([np.nan, np.na, np.inf, -np.inf], 0)
    k, p = stats.normaltest(data, nan_policy='omit')
    k_log, p_log = stats.normaltest(np.log(data[data != 0]), nan_policy='omit')
    #the normality test has a null hypothesis that the data is normally distributed 
    #in this case, any p_val < alpha allows us to reject the null (i.e the distribution is not normal)
    return {'P-value Raw Distribution':p, 
            'P-value Log Distribution': p_log,
            'IsRawNormal': False if p <= alpha else True,
            'IsLogNormal': False if p_log <= alpha else True} #is an inference on the log data an inference on median or mean?
```

```python
def encode(encoder, data, reshape=False):
    if reshape:
        data = np.array(data).reshape(-1,1) #do not reshape for label encoding, reshape for to 1D array for OHE
    return encoder.fit_transform(data)
```

```python
def scale(scaler, data):
    return scaler(data, with_std=True)    
```

```python
def normalize(data):
    data = np.array(data).reshape(-1,1)
    return pp.Normalizer().fit_transform(data)
```

```python
def run_linear_regression(data, target, score='neg_mean_squared_error'):
    Xs = data.drop([target], axis=1)
    y = data[target].values.reshape(-1, 1)
    
    #train test split
    #X_train, X_test, y_train, y_test =train_test_split(Xs,y, test_size=0.3, random_state=3)
    
    #custom time baseD train test split
    X_train, X_test, y_train, y_test = train_test_split(Xs,y, test_size=0.3, random_state=3)
    
    
    #fit to training set
    lin_reg = LinearRegression().fit(X_train, y_train)
    
    train_score = lin_reg.score(X_train,y_train)
    test_score = lin_reg.score(X_test,y_test)
    MSEs = cross_val_score(lin_reg, X_train, y_train, scoring=score, cv=5)
    #lin_reg = lin_reg.fit(Xs, y)
    return MSEs, lin_reg, train_score, test_score, lin_reg.intercept_, lin_reg.coef_
```

```python
def run_linear_regression_by_date(data, target, date_feature, score='neg_mean_squared_error'):
    #Xs = data.drop([target], axis=1)
    #y = data[target].values.reshape(-1, 1)
    
    #train test split
    #X_train, X_test, y_train, y_test =train_test_split(Xs,y, test_size=0.3, random_state=3)
    if data[date_feature].dtype != np.datetime64:
        data[date_feature] = pd.to_datetime(data[date_feature], format='%Y-%m-%d')    #custom time based train test split
    train_start_date = data[date_feature].min()
    end_test_date = data[date_feature].max()
    
    X_train, X_test, y_train, y_test = train_test_split_by_date(data, date_feature, target, train_start_date, dt.datetime.strptime('2018-12-31', '%Y-%m-%d'), dt.datetime.strptime('2019-01-01', '%Y-%m-%d'), end_test_date)
    print(X_train.shape)
    print(len(y_train))
    #fit to training set
    lin_reg = LinearRegression().fit(X_train, y_train)
    
    train_score = lin_reg.score(X_train,y_train)
    test_score = lin_reg.score(X_test,y_test)
    MSEs = cross_val_score(lin_reg, X_train, y_train, scoring=score, cv=5)
    #lin_reg = lin_reg.fit(Xs, y)
    return MSEs, lin_reg, train_score, test_score, lin_reg.intercept_, lin_reg.coef_
```

```python
def train_test_split_by_date(data, date_feature, target_feature, start_train_date, end_train_date, start_test_date, end_test_date):
    train_slice = data[(data[date_feature] >= start_train_date) & (data[date_feature] <= end_train_date)].drop([date_feature], axis=1)
    test_slice = data[(data[date_feature] >= start_test_date) & (data[date_feature] <= end_test_date)].drop([date_feature], axis=1)
    #train_slice.drop(columns=[date_feature], axis=1, inplace=True)
    #test_slice.drop(columns=[date_feature], axis=1, inplace=True)
    y_train = train_slice[target_feature].values.reshape(-1, 1)
    y_test = test_slice[target_feature].values.reshape(-1, 1)
    X_train = train_slice.drop([target_feature], axis=1)
    X_test = test_slice.drop([target_feature], axis=1)
    return X_train, X_test, y_train, y_test
```

```python
def run_ridge_regression(data, target, alphas={'alpha':[1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}):
    Xs = data.drop([target], axis=1)
    y = data[target].values.reshape(-1, 1)
    ridge = Ridge()
    ridge_regression = GridSearchCV(ridge, alphas, scoring='neg_mean_squared_error', cv=5)
    ridge_regression.fit(Xs, y)
    print(ridge_regression.best_params_)
    print(ridge_regression.best_score_)
```

```python
def run_lasso_regression(data, target, alphas={'alpha':[1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}):
    Xs = data.drop([target], axis=1)
    y = data[target].values.reshape(-1, 1)
    lasso = Lasso()
    lasso_regression = GridSearchCV(lasso, alphas, scoring='neg_mean_squared_error', cv=5)
    lasso_regression.fit(Xs, y)
    print(lasso_regression.best_params_)
    print(lasso_regression.best_score_)
```
